```python
import cobra
import numpy as np
import time
```

```python
#### REMEMBER TO SAVE RUN TIMES!!!! 
```

```python
def samples_gen(name,seeed,thns,alph):
    model = cobra.io.read_sbml_model("{}.xml".format(name))
    print('{} model read in'.format(name))
    print("Model has {} reactions and {} metabolites".format(len(model.reactions),len(model.metabolites)))
    n = 5000
    print("Optgp sampling ...")
    start = time.time()
    samples = cobra.flux_analysis.sample(model,n, method='optgp', processes=2, seed=seeed, thinning=thns) 
    #can change number of processes to number of processors on laptop
    end = time.time()
    print(end-start)
    print("Optgp sampler finished")
    savename = "ResultsOPTGP_{}_{}_N5000_T{}.csv".format(name,alph,thns) #save name and location
    np.savetxt(savename, samples, delimiter=",")
    print("ACHR sampling ...")
    start = time.time()
    samples = cobra.flux_analysis.sample(model,n, method='achr',seed=seeed, thinning = thns)
    end = time.time()
    print(end-start)
    print("ACHR sampler finished")
    savename = "ResultsACHR_{}_{}_N5000_T{}.csv".format(name,alph,thns) #save name and location
    np.savetxt(savename, samples, delimiter=",")
```

```python
modls = ["Arnold2014", "Gomes2010","Poolman2009", 
         "Arnold2014", "Gomes2010","Poolman2009", 
         "Arnold2014", "Gomes2010","Poolman2009",
        "Arnold2014", "Gomes2010","Poolman2009",
         "Arnold2014", "Gomes2010","Poolman2009", 
         "Arnold2014", "Gomes2010","Poolman2009"]
seeds = [35,35,35,33,33,33,34,34,34,75,75,75,73,73,73,74,74,74]
#thins = [532512,231200,192200,532512,231200,192200]
thins = [100,100,100,1000,1000,1000,10000,10000,10000,100,100,100,1000,1000,1000,10000,10000,10000,]
alpha = ['B','B','B','B','B','B','B','B','B','C','C','C','C','C','C','C','C','C']

for i in range(7,18):
    samples_gen(modls[i],seeds[i],thins[i],alpha[i])
```
